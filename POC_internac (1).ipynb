```python
aborrar = pepe[['Desde','Cama','Paciente','Aislamiento']].shift(-1)
aborrar = aborrar.rename(columns={'Aislamiento': "Aislamiento_s",'Cama': "Cama_s", "Paciente": "Paciente_s","Desde": "Desde_s"}) 

pepe_aborrar = pd.concat([pepe, aborrar], axis=1).reindex(pepe.index) # Pega directamente "aborrar" a derecha de la "pepe"
len(pepe),len(pepe_aborrar)
```

```python
# Siempre queda un caso missing en la última fila en todos los campos "VAR_i_s" cuando se hace el shift(-1) de la tabla [VAR_1, ..., VAR_n]
# Esto es porque es un desplazamiento hacia arriba de las "VAR_i". Entonces los llenamos copiando los mismos valores que tienen las "VAR_i"

pepe_aborrar['Desde_s'] = pepe_aborrar.apply(lambda x: x['Desde'] if str(x['Desde_s']) == 'nan' else x['Desde_s'], axis=1)
pepe_aborrar['Cama_s'] = pepe_aborrar.apply(lambda x: x['Cama'] if str(x['Cama_s']) == 'nan' else x['Cama_s'], axis=1)
pepe_aborrar['Paciente_s'] = pepe_aborrar.apply(lambda x: x['Paciente'] if str(x['Paciente_s']) == 'nan' else x['Paciente_s'], axis=1)
pepe_aborrar['Aislamiento_s'] = pepe_aborrar.apply(lambda x: x['Aislamiento'] if str(x['Aislamiento_s']) == 'nan' else x['Aislamiento_s'], axis=1)
```

```python
pepe_aborrar['Tipo Egreso'].value_counts(),len(pepe_aborrar)
```

```python
################################################################################################################################
pepe_aborrar['es_pase'] = pepe_aborrar.apply(lambda x: ((datetime.datetime.strptime(str(x['Desde_s']),'%Y-%m-%d %H:%M') - 
                                                     datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M')).days + 1)
                                          if (x['Paciente'] == x['Paciente_s'] and x['Aislamiento'] != x['Aislamiento_s'] and 
                                             ((datetime.datetime.strptime(str(x['Desde_s']),'%Y-%m-%d %H:%M') - 
                                               datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M')).days + 1) == 1) else '', axis=1)
pepe_aborrar['es_pase'] = pepe_aborrar.apply(lambda x: 1 if x['es_pase'] == 1 else 0, axis=1)
#and x['Cama_s'] != x['Cama'] <---- OJO: Pedir que la cama sea distinta es un cambio de cama en la sala pero tal vez NO es un "Pase de Servicio"

################################################################################################################################
pepe_aborrar['es_cambio_cama'] = pepe_aborrar.apply(lambda x: ((datetime.datetime.strptime(str(x['Desde_s']),'%Y-%m-%d %H:%M') - 
                                                     datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M')).days + 1)
                                          if (x['Paciente'] == x['Paciente_s'] and x['Cama'] != x['Cama_s']) else '', axis=1)
# Ojo que "es_cambio_cama" puede ser un "es_pase" como también ser un reingreso

################################################################################################################################
pepe_aborrar['es_otra_internacion_mismo_paciente'] = pepe_aborrar.apply(lambda x: 
                                                                        ((datetime.datetime.strptime(str(x['Desde_s']),'%Y-%m-%d %H:%M') - 
                                                                          datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M')).days + 1) 
                                                                        if (((datetime.datetime.strptime(str(x['Desde_s']),'%Y-%m-%d %H:%M') - 
                                                                              datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M')).days + 1) > 1 
                                                                            and x['Paciente'] == x['Paciente_s']
                                                                           ) 
                                                                        else '', axis=1)
pepe_aborrar['es_otra_internacion_mismo_paciente'] = pepe_aborrar.apply(lambda x: 1 if x['es_otra_internacion_mismo_paciente'] != '' else '', axis=1)
```

```python
pepe_aborrar['pacdias'] = pepe_aborrar.apply(lambda x:
                                                    ((datetime.datetime.strptime(str(x['aux1']),'%Y-%m-%d %H:%M').replace(hour=0,minute=0) - 
                                                      datetime.datetime.strptime(str(x['Desde']),'%Y-%m-%d %H:%M').replace(hour=0,minute=0)).days + 1),
                                                    axis=1)

pepe_aborrar['es_alta'] = pepe_aborrar.apply(lambda x: 1 if x['Tipo Egreso'] == 'Normal' else 0, axis=1)
```

```python
pepe_aborrar['en_actividad'] = pepe_aborrar.apply(lambda x: 1 if (x['es_cambio_cama'] == '' and x['Revista'] == 'Actividad') else 0, axis=1)
pepe_aborrar['es_defuncion'] = pepe_aborrar.apply(lambda x: 1 if x['Tipo Egreso'] == 'Óbito' else 0, axis=1)
pepe_aborrar['es_derivacion'] = pepe_aborrar.apply(lambda x: 1 if x['Tipo Egreso'] == 'Derivacion Externa' else 0, axis=1)
pepe_aborrar['es_internado'] = pepe_aborrar.apply(lambda x: 1 if x['es_cambio_cama'] == '' else 0, axis=1)
pepe_aborrar['es_cuna'] = pepe_aborrar.apply(lambda x: 1 if chequeotexto(str(x['Cama']),'CUNA') == 1 else 0, axis=1)
```

```python
# El egreso de un paciente de un servicio puede ser por PASE o por ALTA. Entonces se tienen los casos:
#    1) Un paciente que extiende su internación al siguiente mes del mes del reporte computa al servicio como INTERNADO = 1 y EGRESO 
#       del servicio = 0.
#    2) Un paciente internado en CLINICA MEDICA que pasa a CIRUGIA GENERAL, luego al quirófano, luego es compensado en UTI y 
#       luego vuelve a CLINICA MEDICA para control y luego egresa de CLINICA MEDICA, todo durante días del mes del reporte, computa a 
#.      éste servicio como INTERNADO = 1 y EGRESO del servicio = 2.

#    CONCLUSION: Un servicio puede tener MAS internados que egresados, IGUAL cantidad, o MENOS.

pepe_aborrar['es_egreso'] = pepe_aborrar.apply(lambda x: 1 
                                               if (x['seguia_internado'] == 0 and (x['es_internado'] == 1 or x['es_pase'] == 1)) 
                                               else 0, axis=1)

tortaBIS = pepe_aborrar.groupby('Paciente', as_index=False).agg({'es_egreso': 'max'})
# OJO CON EL MATCHING:
pepe_aborrar = pepe_aborrar.merge(tortaBIS, how='inner', on='Paciente')
pepe_aborrar = pepe_aborrar.rename(columns={"es_egreso_x": "es_egreso","es_egreso_y": "es_egreso_extendido"}) 

pepe_aborrar['diasestadia'] = pepe_aborrar.apply(lambda x: x['pacdias'] if x['es_egreso_extendido'] == 1 else 0, axis=1)

(pepe_aborrar['es_internado'].sum(), pepe_aborrar['es_egreso'].sum(), pepe_aborrar['en_actividad'].sum(), pepe_aborrar['es_defuncion'].sum(), 
 pepe_aborrar['es_derivacion'].sum(), pepe_aborrar['diasestadia'].sum(), pepe_aborrar['pacdias'].sum())
```
